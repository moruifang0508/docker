1） docker run : 创建一个新的容器并运行一个命令
    
语法：

docker run [options] image [command] [arg...]

实例：

docker run ubuntu:16.04 /bin/echo "hello world"

2) docker start:启动一个或多个已经被停止的容器

docker stop:停止一个运行中的容器

docker restart:重启容器

语法：

docker start [options] container [container]

docker stop [options] container [container]

docker restart [options] container [container]

实例：

docker start ubuntu  #启动被停止的容器ubuntu

docker stop ubuntu #停止运行中的容器ubuntu

docker restart ubuntu #重启容器ubuntu

3）docker kill:杀掉一个或多个运行中的容器
    
语法：

docker kill [options] container [container...]

options说明：

-s：向容器发送一个信号
 
实例：

docker kill -s mynginx #杀掉运行中的容器mynginx


4)docker rm:删除一个或多个容器

语法：

docker rm [options] container [container...]

options说明：

-f：强行删除一个运行中的容器

-l：移除容器间的网络连接，而非容器本身

-v：删除与容器相关的卷

实例：

docker rm -f db01 db02  #强制删除容器db01、db02

docker rm -l db  #移除容器nginx01对容器db01的连接，连接名db

docker rm -v nginx01 #删除容器nginx01，并删除容器挂载的数据卷

5）docker pause:暂停容器中的所有进程

docker unpause:恢复容器中的所有进程

语法：

docker pause [options] container [container...]

docker unpause [options] container [container...]

实例：

docker pause db01 #暂停容器db01的所有进程

docker unpause db01 #恢复容器db01的所有进程

6）docker create:创建一个新的容器但是不起动它，用法同docker run

语法：

docker create [options] image [command] [arg...]

实例：

docker create --name myrunoob nginx:latest #使用镜像ngin:latest创建一个容器，并將容器命名为myrunoob

7） docker exec:在运行的容器中执行命令

语法：

docker exec [options] container command [arg...]

options:

-d:在后台运行

-i：即使没有附加也保持STDIN打开

-t：分配一个伪终端

实例：

docker exec -it myginx /bin/sh /root/runoob.sh #在容器myginx中以交互模式执行容器内/root/runoob.sh脚本

docker exec -i -t myginx /bin/bash #在容器myginx中开启一个交互模式的终端
