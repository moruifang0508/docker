1）docker commit:从容器创建一个新的对象

语法：

docker commit [options] container [repository[:tag]]

options说明：

-a：提交的镜像作者

-c：使用Dockerfile指令来创建镜像

-m：提交时的说明文字

-p：在commit时，將容器暂停


实例：

docker commit -a "llcuen" -m "my apache" a404c6c174a2 mymysql:v1 mymysql:v1 

#將容器a404c6c174a2保存为新的对象，并添加提交人信息和说明信息

docker images mymysql:v1

返回：

REPOSITORY          TAG                 IMAGE ID            CREATED             SIZE

mymysql             v1                  37af1236adef        15 seconds ago      329 MB


2）docker cp：用于容器与主机之间的数据拷贝

语法：

docker cp [options] container:src_path dest_path 

docker cp [options] src_path container:dest_path

options说明:

-L:保持目标中的链接

实例：

docker cp /www/runoob 96f7f14e99ab:/www/ #將主机/www/runoob目录拷贝到容器96f7f14e99ab的/www目录下

docker cp 96f7f14e99ab:/www /tmp/ #將容器96f7f14e99ab的/www目录拷贝到主机的/tmp目录中

3）docker diff：检查容器里文件结构的更改

语法：

docker diff [options] container

实例：

docker diff mymysql  #查看容器mymysql的文件结构更改
